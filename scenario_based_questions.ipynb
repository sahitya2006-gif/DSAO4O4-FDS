```python
import numpy as np

# Columns: Bedrooms, Sqft, Sale_Price
house_data = np.array([
    [3, 1400, 7500000],
    [5, 2200, 12500000],
    [4, 1800, 9800000],
    [6, 3000, 18500000],
    [2, 1100, 5200000]
])

# Houses with more than 4 bedrooms
filtered = house_data[house_data[:, 0] > 4]

# Average sale price
average_price = np.mean(filtered[:, 2])

print("Average Sale Price:", average_price)
```

```python
from sklearn.linear_model import LogisticRegression

# Features
X = [
    [1, 3, 120],
    [6, 15, 300],
    [0, 1, 80],
    [4, 10, 250],
    [2, 5, 160]
]

# Target
y = [0, 1, 0, 1, 0]

# Model
model = LogisticRegression()
model.fit(X, y)

# User Input
links = int(input("Enter Number of Links: "))
caps = int(input("Enter Number of Capitalized Words: "))
length = int(input("Enter Email Length: "))

prediction = model.predict([[links, caps, length]])

if prediction[0] == 1:
    print("Spam Email")
else:
    print("Not Spam")
```

```python
from sklearn.linear_model import LogisticRegression

X = [
    [1, 3, 120],
    [6, 15, 300],
    [0, 1, 80],
    [4, 10, 250],
    [2, 5, 160]
]

y = [0, 1, 0, 1, 0]

model = LogisticRegression()
model.fit(X, y)

new_email = [[3, 8, 200]]

prediction = model.predict(new_email)

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Spam")
else:
    print("Not Spam")
```

```python
from sklearn.linear_model import LinearRegression

# Features: Brand_Code, Age, Kms_Driven, Engine_CC
X = [
    [1, 2, 25000, 1200],
    [2, 5, 60000, 1500],
    [3, 3, 40000, 1300],
    [4, 7, 90000, 1800],
    [5, 1, 15000, 1000]
]

# Target
y = [650000, 420000, 520000, 300000, 720000]

model = LinearRegression()
model.fit(X, y)

print("Brand Codes:")
print("1-Maruti")
print("2-Hyundai")
print("3-Tata")
print("4-Honda")
print("5-Toyota")

brand = int(input("Enter Brand Code: "))
age = int(input("Enter Car Age (Years): "))
kms = int(input("Enter Kilometers Driven: "))
engine = int(input("Enter Engine CC: "))

price = model.predict([[brand, age, kms, engine]])

print("Estimated Resale Price: ₹", round(price[0], 2))
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text

# Features
X = [
    [45000, 720, 0],
    [30000, 650, 1],
    [60000, 780, 0],
    [28000, 620, 2],
    [52000, 710, 1]
]

# Target
y = [1, 0, 1, 0, 1]

# CART Model
model = DecisionTreeClassifier(criterion="gini")
model.fit(X, y)

# User Input
income = int(input("Enter Monthly Income: "))
credit = int(input("Enter Credit Score: "))
loans = int(input("Enter Existing Loans: "))

prediction = model.predict([[income, credit, loans]])

if prediction[0] == 1:
    print("Loan Approved")
else:
    print("Loan Rejected")

# Display Decision Rules
rules = export_text(model, feature_names=["Income", "Credit_Score", "Existing_Loans"])
print("\nDecision Rules:")
print(rules)
```
